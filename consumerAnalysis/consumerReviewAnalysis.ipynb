```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pre_df = pd.read_csv("customerReviewCleaned.csv")
df = pre_df.iloc[0:20]
df
```

```python
mean_sound_importance = df["Sound Quality Importance"].mean()
mean_battery_importance = df["Battery Importance"].mean()
mean_design_importance = df["Design Importance"].mean()
mean_connectivity_importance = df["Connectivity Importance"].mean()
mean_durability_importance = df["Durability Importance"].mean()
mean_price_importance = df["Price Importance"].mean()
```

```python
importance_means = np.array([mean_sound_importance, mean_battery_importance, mean_design_importance, mean_connectivity_importance, mean_durability_importance, mean_price_importance])
categories = np.array(["Sound", "Battery", "Design", "Connectivity", "Durability", "Pricing"])

for i in range(len(importance_means)):
    print("{}: {}".format(categories[i], importance_means[i]))
```

```python
med_sound_importance = df["Sound Quality Importance"].median()
med_battery_importance = df["Battery Importance"].median()
med_design_importance = df["Design Importance"].median()
med_connectivity_importance = df["Connectivity Importance"].median()
med_durability_importance = df["Durability Importance"].median()
med_price_importance = df["Price Importance"].median()

median_importance = np.array([med_sound_importance, med_battery_importance, med_design_importance, med_connectivity_importance, med_durability_importance, med_price_importance])


for i in range(len(median_importance)):
    print("{}: {}".format(categories[i], median_importance[i]))
```

```python
mode_sound_importance = df["Sound Quality Importance"].mode()
mode_battery_importance = df["Battery Importance"].mode()
mode_design_importance = df["Design Importance"].mode()
mode_connectivity_importance = df["Connectivity Importance"].mode()
mode_durability_importance = df["Durability Importance"].mode()
mode_price_importance = df["Price Importance"].mode()

print(mode_sound_importance)
print(mode_battery_importance)
print(mode_design_importance)
print(mode_connectivity_importance)
print(mode_durability_importance)
print(mode_price_importance)
```

## Analysis of important features:
- The most important feature on a speaker when deciding which one to buy is sound quality (Mean: 4.3, Median: 5.0). The least important is between the design of a speaker and the durability of the speaker.

# What convinces customers to buy?
We reviewed several categories including: Recommendation from family/friends, Online Reviews, Expert Reviews, Brand Reputation, Price, Specific Features, and advertisements

Let's find out which of these work the best to move customers to take action:

```python
mean_recommendation_cta = df["Recommendation Action"].mean()
mean_online_review_cta = df["Online Reviews Action"].mean()
mean_expert_review_cta = df["Expert Reviews Action"].mean()
mean_reputation_cta = df["Brand Reputation Action"].mean()
mean_price_cta = df["Price Action"].mean()
mean_features_cta = df["Specific Features Action"].mean()
mean_advertising_cta = df["Advertising Action"].mean()

cta_array = np.array([mean_recommendation_cta, mean_online_review_cta, mean_expert_review_cta, mean_reputation_cta, mean_price_cta, mean_features_cta, mean_advertising_cta])

cta_index = ["Recommendataion", "Online Reviews", "Expert Reviews", "Reputation", "Price", "Features", "Advertising"]

print("Mean effectiveness of Calls to Action:")
print("---------------------------------------")

for i in range(len(cta_index)):
    print("{}: {}".format(cta_index[i], cta_array[i]))
```

```python
median_recommendation_cta = df["Recommendation Action"].median()
median_online_review_cta = df["Online Reviews Action"].median()
median_expert_review_cta = df["Expert Reviews Action"].median()
median_reputation_cta = df["Brand Reputation Action"].median()
median_price_cta = df["Price Action"].median()
median_features_cta = df["Specific Features Action"].median()
median_advertising_cta = df["Advertising Action"].median()

cta_median_array = np.array([median_recommendation_cta, median_online_review_cta, median_expert_review_cta, median_reputation_cta, median_price_cta, median_features_cta, median_advertising_cta])

print("Median effectiveness of Calls to Action:")
print("----------------------------------------")

for i in range(len(cta_index)):
    print("{}: {}".format(cta_index[i], cta_median_array[i]))
```

```python
mode_recommendation_cta = df["Recommendation Action"].mode()
mode_online_review_cta = df["Online Reviews Action"].mode()
mode_expert_review_cta = df["Expert Reviews Action"].mode()
mode_reputation_cta = df["Brand Reputation Action"].mode()
mode_price_cta = df["Price Action"].mode()
mode_features_cta = df["Specific Features Action"].mode()
mode_advertising_cta = df["Advertising Action"].mode()

print(mode_recommendation_cta)
print(mode_online_review_cta)
print(mode_expert_review_cta)
print(mode_reputation_cta)
print(mode_price_cta)
print(mode_features_cta)
print(mode_advertising_cta)
```

# Demographics and Brand Information
Here, we will observe some demographics (gender, payment amount, etc) and brand information:

```python
percentage_gender = df["What is your gender?"].value_counts(normalize = True) * 100
print(percentage_gender)
```

```python
percentage_age = df["How old are you?"].value_counts(normalize = True) * 100
print(percentage_age)
```

```python
percentage_income = df["What's your annual household income?"].value_counts(normalize = True) * 100
print(percentage_income)
```

```python
percentage_brands = df["Brands:"].str.split(",").explode().str.strip().value_counts(normalize = True) * 100
print(percentage_brands)
```

```python
exploded_df = df.assign(Brands=df['Brands:'].str.split(',')).explode('Brands')
exploded_df['Brands'] = exploded_df['Brands'].str.strip()
brands_described = exploded_df.groupby('Brands')['Satisfaction:'].describe()
print(brands_described)
```

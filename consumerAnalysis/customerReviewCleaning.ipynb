```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customerReview.csv')
```

```python
df = df[df["Ownership Status"] != "No"]
```

```python
df["Additional Comments"] = df["Additional Comments"].str.lower()
df["Additional Comments"] = df["Additional Comments"].str.replace("/", "", regex=True)
```

```python
df["Additional Comments"] = df["Additional Comments"].replace("na", np.nan, regex=True)
df
```

```python
df = df.drop("Score", axis = 1)
```

```python
cols = ["Recommendation Action", "Online Reviews Action", "Expert Reviews Action", "Brand Reputation Action", "Price Action", "Specific Features Action", "Advertising Action"]
df[cols] = df[cols].apply(lambda col: col.str.replace(r"\s*\(.*\)", "", regex = True), axis = 1)
```

```python
df["Additional Comments"] = df["Additional Comments"].replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df["Additional Comments"] = df["Additional Comments"].replace('', np.nan, regex = True)
```

```python
df
```

```python
df.to_csv('customerReviewCleaned.csv', index = False)
```
